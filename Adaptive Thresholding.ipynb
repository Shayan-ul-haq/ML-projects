```python
import cv2

# Read image
img = cv2.imread("Coins.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Blur if needed
blur = cv2.GaussianBlur(gray, (5,5), 0)

# Global threshold
_, global_thresh = cv2.threshold(blur, 127, 255, cv2.THRESH_BINARY)

# Adaptive Mean Threshold
adaptive_mean = cv2.adaptiveThreshold(
    blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2
)

# Adaptive Gaussian Threshold
adaptive_gauss = cv2.adaptiveThreshold(
    blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
)

cv2.imshow("Global Threshold", global_thresh)
cv2.imshow("Adaptive Mean", adaptive_mean)
cv2.imshow("Adaptive Gaussian", adaptive_gauss)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
