```python
import cv2
import numpy as np

# ------------------------------
# Load image (grayscale)
# ------------------------------
img = cv2.imread('img.tif', cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Error: Could not load image. Check that 'img.tif' is in the same folder.")
    exit()

cv2.imshow("Original", img)
cv2.waitKey(0)

# ------------------------------
# 1. Negative Transform
# ------------------------------
negative = 255 - img

cv2.imshow("Negative", negative)
cv2.waitKey(0)

# ------------------------------
# 2. Log Transform
# ------------------------------
img_float = img.astype(np.float32)
c = 255 / np.log(1 + np.max(img_float))   # normalization constant
log_image = c * np.log(1 + img_float)
log_image = np.uint8(log_image)

cv2.imshow("Log Transform", log_image)
cv2.waitKey(0)

# ------------------------------
# 3. Gamma Transform
# ------------------------------
gamma = 0.5    # you can change this to any value
gamma_image = np.array(255 * (img_float / 255) ** gamma, dtype=np.uint8)

cv2.imshow(f"Gamma Transform (gamma={gamma})", gamma_image)
cv2.waitKey(0)

cv2.destroyAllWindows()
```
